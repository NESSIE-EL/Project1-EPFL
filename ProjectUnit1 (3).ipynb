```python
def matrice (lo):
  
    for i in range(lo.shape[0]):
        for j in range(lo.shape[1]):
            if lo.iloc[i,j]!=0.0:
                lo.iloc[i,j]=1
    return lo
#Ici l'objectif est de crer une fonction qui modifie les valeurs de la matrice ici si on a une valeur non nulle on la transforme en 1
```
